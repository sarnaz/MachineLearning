```python
import csv

def _load_csv( filename, hasHeader = True ) : 
  headers = None
  data = list()
  with open( filename, 'r' ) as inhandle: 
    csvfile = csv.reader( inhandle ) 
    for row in csvfile : 
      if hasHeader : 
        headers = row 
        hasHeader = False
        continue
      data.append( row ) 
  return headers, data 
```

```python
headers, data = _load_csv( 'train.csv' ) 
print( list( enumerate( headers ) ) ) 
```
